วาดกราฟ parallel coordinate เปรียบเทียบ [อายุ และ เพศ] โดยใช้ class เป็น risks (เลือกมาแค่ 3 risks)

```python
import pandas as pd
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/Midterm'
```

```python
data_file_path = os.path.join(path,'confirmed-cases.csv') 
data_df = pd.read_csv(data_file_path)
```

```python
data_df #นำข้อมูลมาจาก drive
```

```python
data_h = data_df[['sex','age','risk']] #เลือกเอาเฉพาะข้อมูลที่ต้องการใช้
data_h
```

```python
data_h.isnull().any() #ตรวจสอบค่าMissing
```

```python
data_risk = data_h.dropna() #แก้Missingด้วยการ Dropan
data_risk
```

```python
data_risk.isnull().any() #ตรวจสอบค่าMissingอีกครั้ง
```

```python
foreign = data_risk[data_risk['risk']=='State Quarantine'] #เลือกเอาเฉพาะState Quarantine
foreign
```

```python
investigation = data_risk[data_risk['risk']=='สนามมวย'] #เลือกเอาเฉพาะสนามมวย
investigation
```

```python
other = data_risk[data_risk['risk']=='อาชีพเสี่ยง เช่น ทำงานในสถานที่แออัด หรือทำงานใกล้ชิดสัมผัสชาวต่างชาติ เป็นต้น'] #เลือกเอาเฉพาะอาชีพเสี่ยง เช่น ทำงานในสถานที่แออัด หรือทำงานใกล้ชิดสัมผัสชาวต่างชาติ เป็นต้น
other
```

```python
data = pd.concat([foreign,investigation,other]) #รวมตาราง
data
```

```python
data_k = data.replace('หญิง',1) #กำหนดเพศหญิงเป็น1
data_k
```

```python
data_final = data_k.replace('ชาย',100) #กำหนดเพศชายเป็น100
data_final
```

```python
data_y = data_final[['age','risk']] #เลือกเอาเฉพาะข้อมูลที่จะทำกราฟ
data_y
```

```python
data_final['risk'].unique() #เลือกดูชื่อrisk
```

```python
pd.plotting.parallel_coordinates(data_final, 'risk'); #พอตกราฟ
```
